```python
key = [[0,0,0], 
       [1,0,0], 
       [0,1,1]]

lock = [[1,1,1],
       [1,1,0],
       [1,0,1]]
```

```python
def minimize_square(lock_data):
    col_min = len(lock_data)
    col_max = 0
    row_min = len(lock_data)
    row_max = 0
    
    
    for i in range(len(lock_data)):
        for j in range(len(lock_data)):
            if lock[i][j] == 0:
                col_min = min(col_min, j)
                col_max = max(col_max, j)
                row_min = min(row_min, i)
                row_max = max(row_max, i)
                
    return row_min, row_max, col_min, col_max


a,b,c,d = minimize_square(lock)

import numpy as np
lock = np.array(lock)
key = np.array(key)

minimize_lock = lock[a:b+1, c:d+1]


# key, minimize_lock

def check_all_pos(key, minimize_lock):
    column_length = len(minimize_lock[0])
    row_length = len(minimize_lock)
    number = column_length * row_length

    if len(key) > row_length and len(key[0]) > column_length:
        for i in range(len(key)-row_length+1): #  
            for j in range(len(key[0]) - column_length + 1): # 
                tmp = key[i:i+row_length, j:j+column_length]
                tmp1 = tmp +  minimize_lock
                if np.std(tmp1) == 0 and sum(tmp1) == number:
                    return True
                else:
                    continue
        return False
    else:
        return False
    
    
def rotate_clock(mat):
    '''  '''
    return np.rot90(mat, axes = (1,0))


for i in range(4):
    answer = check_all_pos(key, minimize_lock)
    key = rotate_clock(key)
    if answer == True:
        break
        
        
answer
```
