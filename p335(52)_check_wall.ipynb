```python
def distance_bw_walls(walls, n):
    distance_bw = []
    walls = sorted(walls)
    distance_bw.append(walls[0]+n - walls[-1])
    
    for i in range(1,len(walls)):
        distance_bw.append(abs(walls[i] - walls[i-1]))
    return distance_bw


def pick_max_worker(worker):
    '''pop and return max worker'''
    max_num_idx = [idx for idx, x in enumerate(worker) if x == max(worker)]
    max_num = worker.pop(max_num_idx[0])
    return max_num

kk = distance_bw_walls(weak, n)

# find max, and cut that part, make circle to line
max_kk = [[x, idx] for idx, x in enumerate(kk) if x == max(kk)]

# cut and rearrange
cnt_array = []
for i in range(len(max_kk)):
    marker = max_kk[i][1]
    
    end = weak[marker:]
    for x in weak[:marker]:
        end.append(x+12)

    #distance bw walls after cutting
    final_distance = distance_bw_walls(end,n)[1:]

    cnt = 0
    while len(dist) > 0:
        tmp_max = max(dist)
        dist.remove(tmp_max)
        tmp_summation = 0
        while tmp_summation <= tmp_max and len(final_distance) > 0:
            tmp_summation += final_distance.pop(0)

        cnt += 1
        if len(final_distance) == 0:
            break
            
    if len(final_distance) == 0:
        answer = cnt
    elif len(dist) == 0:
        answer = -1
        
    cnt_array.append(answer)

if max(cnt_array) > 0:
    while True:
        try:
            cnt_array.remove(-1)
        except:
            break
    answer = min(cnt_array)    
else:
    answer = -1
```
