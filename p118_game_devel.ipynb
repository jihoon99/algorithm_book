```python
# map size
map_size = [4,4]
# character location and direction
lo_di = [1,1,0] # locations and direction - 0 : North, 1: East, 2: South, 3: West
map_info = [    # 0 : lands, 1 : sea which cant stand on 
    [1,1,1,1],
    [1,0,0,1],
    [1,1,0,1],
    [1,1,1,1]
]

# return : number of places visited
```

```python
def movement(location, data, arrow):
    if arrow == 0: # north
        try:
            tmp_location = [location[0] - 1, location[1]] # move to left
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 3 # left
                return tmp_location, arrow
        except:
            pass



        try:
            tmp_location = [location[0], location[1]+1] # move to down
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 2 # down
                return tmp_location, arrow
        except:
            pass

        try:
            tmp_location = [location[0] + 1 , location[1]] # move to right
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 1 # right
                return tmp_location, arrow
        except:
            pass

        try:
            tmp_location = [location[0], location[1] + 1] # move to up
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 0 # up
                return tmp_location, arrow
        except:
            pass
        
    if arrow == 1: # east
        try:
            tmp_location = [location[0], location[1] + 1] # move to up
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 0 # up
                return tmp_location, arrow
        except:
            pass
        
        try:
            tmp_location = [location[0] - 1, location[1]] # move to left
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 3 # left
                return tmp_location, arrow
        except:
            pass


        try:
            tmp_location = [location[0], location[1]+1] # move to down
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 2 # down
                return tmp_location, arrow
        except:
            pass

        try:
            tmp_location = [location[0] + 1 , location[1]] # move to right
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 1 # right
                return tmp_location, arrow
        except:
            pass

    if arrow == 2: # south
        try:
            tmp_location = [location[0] + 1 , location[1]] # move to right
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 1 # right
                return tmp_location, arrow
        except:
            pass
        
        try:
            tmp_location = [location[0], location[1] + 1] # move to up
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 0 # up
                return tmp_location, arrow
        except:
            pass
        
        try:
            tmp_location = [location[0] - 1, location[1]] # move to left
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 3 # left
                return tmp_location, arrow
        except:
            pass


        try:
            tmp_location = [location[0], location[1]+1] # move to down
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 2 # down
                return tmp_location, arrow
        except:
            pass


    if arrow == 3: # west
        try:
            tmp_location = [location[0], location[1]+1] # move to down
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 2 # down
                return tmp_location, arrow
        except:
            pass
        
        try:
            tmp_location = [location[0] + 1 , location[1]] # move to right
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 1 # right
                return tmp_location, arrow
        except:
            pass
        
        try:
            tmp_location = [location[0], location[1] + 1] # move to up
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 0 # up
                return tmp_location, arrow
        except:
            pass
        
        try:
            tmp_location = [location[0] - 1, location[1]] # move to left
            if data[tmp_location[0]][tmp_location[1]] == 0:
                arrow = 3 # left
                return tmp_location, arrow
        except:
            pass


    return None, None

```

```python
cnt = 1
map_info[lo_di[0]][lo_di[1]] = 1

while True:
    try:
        location, direction = movement(lo_di[:2], map_info, lo_di[2])
        lo_di[0], lo_di[1], lo_di[2] = location[0], location[1], direction
        map_info[lo_di[0]][lo_di[1]] = 1
        cnt += 1
    except:
        break
```

# code more efficiently

```python
# map size
map_size = [4,4]
# character location and direction
lo_di = [1,1,0] # [1,1] : locations and // direction '0' which means   0 : North, 1: East, 2: South, 3: West
map_info = [    # 0 : lands, 1 : sea which cant stand on 
    [1,1,1,1],
    [1,0,0,1],
    [1,1,0,1],
    [1,1,1,1]
]

# return : number of places visited
```

```python
# 북, 동, 남, 서 
dx = [-1,0,1,0]
dy = [0,1,0,-1]
direction = lo_di[2]

# turn left
def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3

x = lo_di[0]
y = lo_di[1]

# start simulation
count = 1
turn_time = 0
while True:
    turn_left()
    nx = x + dx[direction]
    ny = y + dy[direction]
        
    # move forward when 0 is in front
    if map_info[nx][ny] == 0:
        map_info[nx][ny] = 1
        x = nx
        y = ny
        count += 1
        turn_time = 0
    else:
        turn_time += 1
    
    if turn_time == 4:
        nx = x -dx[direction]
        ny = y - dy[direction]
        
        if map_info[nx][ny] == 0:
            x = nx
            y = ny
        else:
            break
        turn_time = 0
```
