```python
def solution(a):
    
    if len(a) < 3:
        return 2
    else:
        answer = 2
        for i in range(1,len(a)-1):
            output = 1



            min_left = min(a[:i])
            min_right = min(a[i+1:])
            num = a[i]



            if min_left> num:
                output *= 0
            else:
                output *= 1

            if min_right > num:
                output *= 0
            else:
                output *= 1


            if sum(output) != 1:
                answer += 1
        return answer
```
