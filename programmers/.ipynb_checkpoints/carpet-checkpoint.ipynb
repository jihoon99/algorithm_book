```python
total = brown + yellow
```

```python
container = [] # 후보군
for i in range(3, total//2+1):
    if total % i == 0:
        if int(total/i) >= 3:
            container += [[i, int(total/i)]]
#     if total / 2 < i:
#         break
```

```python
container1 = [[x-2,y-2] for x,y in container if x >= y]
```

```python
answer = [[x+2,y+2] for x,y in container1 if x*y == yellow]
```
