```python
from itertools import combinations

def self_mult(values):
    # 
        result = 1
        for i in values:
            result *= i
        return result

def total(clothes):
    # clothes  
    
    dic = dict()
    for i,j in clothes:
        try:
            dic[j] = dic[j] + 1
        except:
            dic[j] = 1
            
    if sum(dic.values()) == len(dic):
        return True
    
    

    
def solution(clothes):

    if total(clothes):
        #   
        return 2**len(clothes) - 1
    
    
    dic = dict()
    for i,j in clothes:

        try:
            dic[j] +=1

        except:
            dic[j] = 1

    dic_values = list(dic.values())  

    answer = sum(dic_values) #    ,..  /  ///

    for i in range(2,len(dic_values)+1):
        kk = list(combinations(dic_values,i)) # 2,3,4,5   .//   
        for j in kk:
            answer += self_mult(j)

    return answer
```
