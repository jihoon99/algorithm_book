```python
phone_book = ["119",  "1195524421","97674223"]
```

```python
test =[[len(i), idx] for idx,i in enumerate(phone_book)]
test = sorted(test)
dic = dict()

for idx, (i,index) in enumerate(test):
    dic[idx] = phone_book[index]
    
answer = True
length = len(dic)
cnt = 0
while dic and answer == True:
    min_ = dic[cnt]
    min_length = len(min_)
    
    for i in range(cnt+1, length):
        if dic[i][:min_length] == min_:
            answer = False
            break
    
    cnt += 1
    
```
