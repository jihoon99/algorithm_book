```python
def closer(data):
    result = ''
    for x in list(data):
        result += x
        
    return result
```

```python
def find_max(ls, max_num):
    
    if max_num > 1:
        answer = []
        for i,j in ls.items(): # ls[2]
            if j == max_num:
                answer.append(i)

        return answer
    else:
        pass
```

```python
answer = []


for i in course:
    print(i)
    try:
        tmp_answer = max(ls[i], key=lambda k: ls[i][k])
        max_num = ls[i][tmp_answer]
        if max_num > 1:
            candidate = find_max(ls[i], max_num)
            for i in candidate:
    #             print(i)
                answer.append(closer(i))
    except:
        pass
answer = sorted(answer)
```
