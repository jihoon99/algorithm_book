```python
def solution(dartResult):

    ls = [0]

    for idx,i in enumerate(dartResult):
        tmp_result = 0

        if i in ['S','D','T']:
            if i == 'S':
                upper = 1
            elif i == 'D':
                upper = 2
            elif i == 'T':
                upper = 3




            if dartResult[idx-2:idx] == '10' and idx-2 >= 0:
                tmp_result = 10**upper
            else:
                tmp_result = int(dartResult[idx-1])**upper



            try:
                if dartResult[idx+1] in ['*','#']:
                    option = dartResult[idx+1]

            except:
                option = None

            if option == '*':
                tmp_result *= 2
                ls[-1] = ls[-1]*2

            elif option == '#':
                tmp_result *= -1


            ls.append(tmp_result)

    return sum(ls)
```

```python
def solution(dartResult):
    location = [idx for idx, i in enumerate(dartResult) if i in ['S','D','T']]
    location_info =''
    for i in location:
        location_info += dartResult[i]


    ls = []

    for idx,i in enumerate(location):
        left = 'off'
        right = 'off'
        if dartResult[i-1] == '0':
            if dartResult[i-2] == '1':
                left = 'on' # 2개를 인덱싱 해야함.
        try:
            if dartResult[i+1] in ['*','#']:
                right = 'on'
        except:
            right ='off'


        if left == 'on':
            left_index = i-2
        else:
            left_index = i-1

        if right == 'on':
            right_index = i+1
        else:
            right_index = i

        ls.append((left_index,right_index+1, location_info[idx]))


    di = dict()
    di['S'] = 1
    di['D'] = 2
    di['T'] = 3



    result = [] # numbers will be added


    for i,j,k in ls:
        try:
            tmp = dartResult[i:j]
        except:
            tmp = dartResult[i:j-1]


        tmp_location =[zdx for zdx, z in enumerate(tmp) if z == k][0]
        # tmp_length = len(tmp)

        tmp_number = int(tmp[:tmp_location])
        tmp_state = tmp[tmp_location]
        try:
            tmp_bonus = tmp[tmp_location+1:]
        except:
            tmp_bonus = ''

        try:
            if tmp_bonus == '*':
                multiply = 2
                add_ = result[-1]
            elif tmp_bonus == '#':
                multiply = -1
                add_ = 0
            else:
                multiply = 1
                add_ = 0    
        except:
            if tmp_bonus == '*':
                multiply = 2
                add_ = 0
            elif tmp_bonus == '#':
                multiply = -1
                add_ = 0
            else:
                multiply = 1
                add_ = 0   

        # print(f'tmp:{tmp}  tmp_number:{tmp_number}    tmp_bonus:{tmp_bonus}   multiply:{multiply}    add:{add}   di[k]:{di[k]}')
        result.append(multiply*tmp_number**(di[k]) + add_)


    answer = sum(result)
    return answer
```
